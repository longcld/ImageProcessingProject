## Import package

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (20, 20)
```

## Load Image

```python
img = cv2.imread('../data/lane.jpg')
IMAGE_H = 800
IMAGE_W = 1280
img = cv2.resize(img, (IMAGE_W, IMAGE_H))
```

## Transform image

```python
src = np.float32([[0, IMAGE_H], [IMAGE_W, IMAGE_H], [0, 0], [IMAGE_W, 0]])
dst = np.float32([[550, IMAGE_H], [711, IMAGE_H], [0, 0], [IMAGE_W, 0]])

M = cv2.getPerspectiveTransform(src, dst)

img = img[400:IMAGE_H, :]
warped_img = cv2.warpPerspective(img, M, (IMAGE_W, IMAGE_H - 150))
```

## Show result

```python
plt.subplot(121),plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.title('Original Image')
plt.subplot(122),plt.imshow(cv2.cvtColor(warped_img, cv2.COLOR_BGR2RGB)),plt.title('Transformed Image')
plt.show()
```
