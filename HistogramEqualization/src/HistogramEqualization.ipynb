## Load using package

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (12, 12)
```

## Compute Histogram

```python
def compute_hist(img):
    hist = np.zeros((256,), np.uint8)
    h, w = img.shape[:2]
    for i in range(h):
        for j in range(w):
            hist[img[i][j]] += 1
    return hist
```

## Equalize Histogram

```python
def equal_hist(hist):
    cumulator = np.zeros_like(hist, np.float64)
    for i in range(len(cumulator)):
        cumulator[i] = hist[:i].sum()
    new_hist = (cumulator - cumulator.min())/(cumulator.max() - cumulator.min()) * 255
    new_hist = np.uint8(new_hist)
    return new_hist
```

## Apply on test image

```python
img = cv2.imread('../data/test.png')
hist = compute_hist(img).ravel()
new_hist = equal_hist(hist)

new_img = img.copy()
h, w = new_img.shape[:2]
for i in range(h):
    for j in range(w):
        new_img[i,j] = new_hist[new_img[i,j]]
```

## Show result

```python
fig = plt.figure()
ax = plt.subplot(221)
plt.imshow(img)
plt.title("Original Image")
plt.subplot(222)
plt.hist(hist)
plt.title("Original Histogram")

plt.subplot(223)
plt.imshow(new_img)
plt.title("Result Image")
plt.subplot(224)
plt.hist(new_hist)
plt.title("Equation Histogram")
plt.show()
```
