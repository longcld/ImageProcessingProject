```python
from PIL import Image, ImageDraw, ImageFont
import os

image_path_input = '../data/'
image_path_output = '../data/'
image_name_input = 'test.jpg'

im = Image.open(image_path_input + image_name_input)
origin_img = im.copy()
print('Input file size   : ', im.size )
print('Input file name   : ', image_name_input )
print('Input Image Size  : ', os.path.getsize (image_path_input  + image_name_input))

image_name_output = 'test_compressed.jpeg'

im.save(image_path_output + image_name_output ,optimize=True,quality=50) 

print('Output file size  : ', im.size )
print('Output file name  : ', image_name_output)
print('Output Image Size : ', os.path.getsize (image_path_output + image_name_output))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

origin_img = cv2.imread("../data/test.jpg")
compressed_img = cv2.imread("../data/test_compressed.jpeg")

plt.rcParams["figure.figsize"] = (9, 9)
plt.subplot(121),plt.imshow(cv2.cvtColor(origin_img, cv2.COLOR_BGR2RGB)),plt.title('Original Image')
plt.subplot(122),plt.imshow(cv2.cvtColor(compressed_img, cv2.COLOR_BGR2RGB)),plt.title('Compressed Image')
plt.show()
```
