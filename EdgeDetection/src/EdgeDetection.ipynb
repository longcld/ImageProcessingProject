## Import package

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("../data/ball.png", 0)
plt.rcParams["figure.figsize"] = (8, 8)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

## Sobel filter

```python
img_sobelx = cv2.Sobel(img, cv2.CV_8U, 1, 0,ksize=5)
img_sobely = cv2.Sobel(img, cv2.CV_8U, 0, 1,ksize=5)
img_sobel = img_sobelx + img_sobely

plt.rcParams["figure.figsize"] = (20, 20)
plt.subplot(131),plt.imshow(cv2.cvtColor(img_sobelx, cv2.COLOR_BGR2RGB)),plt.title('Sobel X filtered'), plt.axis('off')
plt.subplot(132),plt.imshow(cv2.cvtColor(img_sobely, cv2.COLOR_BGR2RGB)),plt.title('Sobel Y filtered'), plt.axis('off')
plt.subplot(133),plt.imshow(cv2.cvtColor(img_sobel, cv2.COLOR_BGR2RGB)),plt.title('Sobel Filtered'), plt.axis('off')
plt.show()
```

## Prewitt Filter

```python
kernelx = np.array([[1, 1, 1],
                    [0, 0, 0],
                    [-1, -1, -1]])
kernely = np.array([[-1,0,1],
                    [-1,0,1],
                    [-1,0,1]])

img_prewittx = cv2.filter2D(img, -1, kernelx)
img_prewitty = cv2.filter2D(img, -1, kernely)

img_prewitt = img_prewittx + img_prewitty

plt.rcParams["figure.figsize"] = (20, 20)
plt.subplot(131),plt.imshow(cv2.cvtColor(img_prewittx, cv2.COLOR_BGR2RGB)),plt.title('Prewitt X filtered'), plt.axis('off')
plt.subplot(132),plt.imshow(cv2.cvtColor(img_prewitty, cv2.COLOR_BGR2RGB)),plt.title('Prewitt Y filtered'), plt.axis('off')
plt.subplot(133),plt.imshow(cv2.cvtColor(img_prewitt, cv2.COLOR_BGR2RGB)),plt.title('Prewitt Filtered'), plt.axis('off')
plt.show()
```

## Robert Filtered

```python
kernelx = np.array([[-1, 0],
                    [0, 1]])
kernely = np.array([[0, -1],
                    [1, 0]])

img_robertx = cv2.filter2D(img, -1, kernelx)
img_roberty = cv2.filter2D(img, -1, kernely)

img_robert = img_robertx + img_roberty

plt.rcParams["figure.figsize"] = (20, 20)
plt.subplot(131),plt.imshow(cv2.cvtColor(img_robertx, cv2.COLOR_BGR2RGB)),plt.title('Robert X filtered'), plt.axis('off')
plt.subplot(132),plt.imshow(cv2.cvtColor(img_roberty, cv2.COLOR_BGR2RGB)),plt.title('Robert Y filtered'), plt.axis('off')
plt.subplot(133),plt.imshow(cv2.cvtColor(img_robert, cv2.COLOR_BGR2RGB)),plt.title('Robert Filtered'), plt.axis('off')
plt.show()
```

## Laplacian Filter

```python
plt.rcParams["figure.figsize"] = (8, 8)
```

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F)
plt.imshow(laplacian, cmap = 'binary')
plt.axis('off')
plt.show()
```

## Canny

```python
img_canny = cv2.Canny(img, 100, 200)

plt.imshow(cv2.cvtColor(img_canny, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
