## Import package

```python
import cv2
import os
import glob
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (10,10)
```

```python
images = glob.glob('../data/*jpg')
```

## Erosion and Dilation

```python
img = cv2.imread(images[0])
img = cv2.resize(img, (300, 300))
res = img.copy()

gray = cv2.cvtColor(res, cv2.COLOR_BGR2GRAY)

lower = np.array([52, 80, 0])
upper = np.array([180, 255, 255])
hsv = cv2.cvtColor(res, cv2.COLOR_BGR2HSV)

mask_ero = cv2.inRange(hsv, lower, upper)

lower_red = np.array([169, 191, 0])
upper_red = np.array([180, 255, 255])

lower_blue = np.array([96, 105, 84])
upper_blue = np.array([160, 255, 255])

mask_blue = cv2.inRange(hsv, lower_blue, upper_blue)
mask_red = cv2.inRange(hsv, lower_red, upper_red)

mask_dila = cv2.bitwise_or(mask_blue, mask_red)


kernel = np.ones((6,6),np.uint8)
erosion = cv2.erode(mask_ero,kernel,iterations = 1)
kernel = np.ones((9,9),np.uint8)
dilation = cv2.dilate(mask_dila,kernel,iterations = 1)

plt.rcParams["figure.figsize"] = (15, 15)
plt.subplot(221),plt.imshow(cv2.cvtColor(mask_ero, cv2.COLOR_BGR2RGB)),plt.title('Origin Mask'), plt.axis('off')
plt.subplot(222),plt.imshow(cv2.cvtColor(erosion, cv2.COLOR_BGR2RGB)),plt.title('Erosion'), plt.axis('off')
plt.subplot(223),plt.imshow(cv2.cvtColor(mask_dila, cv2.COLOR_BGR2RGB)),plt.title('Origin Mask'), plt.axis('off')
plt.subplot(224),plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_BGR2RGB)),plt.title('Dilation'), plt.axis('off')
plt.show()
```

## Opening and Closing Image

```python
kernel = np.ones((6, 6),np.uint8)
opening = cv2.morphologyEx(mask_ero, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(mask_ero, cv2.MORPH_CLOSE, kernel)

plt.subplot(131),plt.imshow(cv2.cvtColor(mask_ero, cv2.COLOR_BGR2RGB)),plt.title('Origin Mask'), plt.axis('off')
plt.subplot(132),plt.imshow(cv2.cvtColor(opening, cv2.COLOR_BGR2RGB)),plt.title('Opening'), plt.axis('off')
plt.subplot(133),plt.imshow(cv2.cvtColor(closing, cv2.COLOR_BGR2RGB)),plt.title('Closing'), plt.axis('off')
plt.show()
```

## Edge Detection

```python
kernel = np.ones((5,5),np.uint8)
erosion_edge = cv2.erode(dilation,kernel,iterations = 1)

plt.subplot(131),plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_BGR2RGB)),plt.title('Origin Mask'), plt.axis('off')
plt.subplot(132),plt.imshow(cv2.cvtColor(erosion_edge, cv2.COLOR_BGR2RGB)),plt.title('Erosion Mask'), plt.axis('off')
plt.subplot(133),plt.imshow(cv2.cvtColor(dilation - erosion_edge, cv2.COLOR_BGR2RGB)),plt.title('Edge Detected'), plt.axis('off')
plt.show()
```
